```python
import numpy as np
import pandas as pd
df=pd.read_excel('/content/pdf_text_with_coordinates.xlsx')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
print(df.columns)
```

```python
df.drop('Page', inplace=True, axis=1)
```

```python
df
```

```python
df['Text']
```

```python
df = df.reset_index(drop=True)
```

```python
sid.polarity_scores(df.loc[0]['Text'])
```

```python
df['score']=df['Text'].apply(lambda text: sid.polarity_scores(text))
df.head()
```

```python
df['polarity_score'] = df['score'].apply(lambda score_dict: score_dict['compound'])
```

```python
df['sentiment'] = df['polarity_score'].apply(
    lambda c: 'positive' if c > 0 else ('negative' if c < 0 else 'neutral')
)
df.head()
```

```python
selected_df = df[['Text', 'X','Y', 'polarity_score','sentiment']]
```
